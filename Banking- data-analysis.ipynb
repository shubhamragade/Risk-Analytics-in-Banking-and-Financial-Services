```python
X = ['Client ID', 'Name', 'Joined Bank', 'Banking Contact', 'Nationality',
     'Occupation', 'Fee Structure', 'Loyalty Classification']

data = data.drop(X, axis=1)
```

```python
data.head()
```

```python
# print(data.dtypes)

if all(data.dtypes.apply(lambda x : x in ['int64', 'float64'])):
    print("all is numerical form")
else:
    print("some nurical columns still remain")
```

```python
# sale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler = scaler.fit_transform(data)
scaler
```

```python
# build the model
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(scaler)
```

```python
data['cluster'] = kmeans.labels_
```

```python
data.tail()
```

```python
print(data['cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaler)

plt.figure(figsize=(8,6))
plt.scatter(reduced_data[:,0], reduced_data[:,1], c=data['cluster'], cmap='viridis')
plt.title('KMeans Clustering Visualization (2D PCA)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# List to store the inertia for each k
inertia = []

# Try k from 1 to 10
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaler)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method: Optimal k for KMeans')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.xticks(K_range)
plt.show()
```
